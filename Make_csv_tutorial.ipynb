In this notebook, we will explore how to make a folder of txt files into a single csv file. We do this becasue it saves time for colab to load our data everytime.




```
import os
import pandas as pd
```

```
#First we have to give google permission to mount our drive. Click the short link below after you run this cell.
#This link will take you to your google account, select your account, then copy whatever the webpage asks you to copy.
#Paste it below
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```
#Next we find the directory of the folder in which you store all your txt files in the drive
file_directory = os.listdir('/content/drive/My Drive/AWCA/TXTs/Secondary Sources') #Here we use an example in our drive, you should replace this with your path
```

```
Title = []
Content = []
for file in file_directory:
  if file.endswith('.txt'):
    Title.append(file)
    Content.append(open('/content/drive/My Drive/AWCA/TXTs/Secondary Sources/' + file, 'r').read()) ##You still need to replace the path here
```

```
your_csv = pd.DataFrame({'Title':Title, 'Content':Content})
```
